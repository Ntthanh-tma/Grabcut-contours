```python
import numpy as np
import sys
import os
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Steve-jobs.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
mask = np.zeros(img.shape[:2], np.uint8)
#tạo 1 bức ảnh khác có shape là 
bgdModel = np.zeros((1, 65), np.float64)
#background model  == default
#foreground model == default
fgdModel = np.zeros((1, 65), np.float64)

rect = (2,2,630,930)
#rectangle == hinh vuong
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_RECT)
# call grabcut
mask2 = np.where((mask==2)|(mask==0), 0,1).astype('uint8')
#mask2 is the foreground mask. so you only use foreground 
#pixels which is done by setting every pixel 0 that is background, else 1
img = img*mask2[:,:, np.newaxis]
# create an iamge with co-ordinate include x, y BGR_to_RGB
```

```python
imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,127,255,0)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
plt.imshow(imgray)
plt.show()
```
